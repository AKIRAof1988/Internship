```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def visualize_feature_pairs(df):
    numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
    categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

    # Pairing numerical columns and visualizing them row-wise
    for i in range(0, len(numerical_cols), 2):
        plt.figure(figsize=(12, 6))
        
        # Plotting the first numerical feature in the pair
        plt.subplot(1, 2, 1)
        sns.histplot(df[numerical_cols[i]], kde=True, bins=20, color='skyblue')
        plt.title(f'Histogram for {numerical_cols[i]}')
        plt.xlabel(numerical_cols[i])
        plt.ylabel('Frequency')

        # Plotting the second numerical feature in the pair if it exists
        if i + 1 < len(numerical_cols):
            plt.subplot(1, 2, 2)
            sns.histplot(df[numerical_cols[i + 1]], kde=True, bins=20, color='salmon')
            plt.title(f'Histogram for {numerical_cols[i + 1]}')
            plt.xlabel(numerical_cols[i + 1])
            plt.ylabel('Frequency')
        
        plt.tight_layout()
        plt.show()

    # Pairing categorical columns and visualizing them row-wise
    for i in range(0, len(categorical_cols), 2):
        plt.figure(figsize=(12, 6))
        
        # Plotting the first categorical feature in the pair
        plt.subplot(1, 2, 1)
        sns.countplot(data=df, x=categorical_cols[i], palette='pastel')
        plt.title(f'Count Plot for {categorical_cols[i]}')
        plt.xlabel(categorical_cols[i])
        plt.ylabel('Count')
        plt.xticks(rotation=45)

        # Plotting the second categorical feature in the pair if it exists
        if i + 1 < len(categorical_cols):
            plt.subplot(1, 2, 2)
            sns.countplot(data=df, x=categorical_cols[i + 1], palette='pastel')
            plt.title(f'Count Plot for {categorical_cols[i + 1]}')
            plt.xlabel(categorical_cols[i + 1])
            plt.ylabel('Count')
            plt.xticks(rotation=45)
        
        plt.tight_layout()
        plt.show()
        plt.savefig('Distributioffgfgn of features.png')
  

# Visualize feature pairs side by side row-wise
visualize_feature_pairs(df)
#plt.savefig('Distribution of features.png')
```

```python
import pandas as pd

# Assuming 'data' is your DataFrame containing the baseball dataset

# Define the features (X) and the target variable (y)
X = data.drop('W', axis=1)  # Exclude the 'W' column (number of wins) from features
y = data['W']  # Target variable is the 'W' column

# Display shapes to confirm the split
print(f"Shape of X (features): {X.shape}")
print(f"Shape of y (target variable): {y.shape}")
```

```python
from sklearn.model_selection import train_test_split
```

```python
# To train and evaluate a model, let's first split the data into
# train and test data:

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=0,
)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
linreg = LinearRegression()

# Train the model.
linreg.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
# Make predictions on the train set and calculate
# the mean squared error.
print('Train set')
pred = linreg.predict(X_train)
print('Linear Regression mse: {}'.format(mean_squared_error(y_train, pred)))

# Make predictions on the test set and calculate
# the mean squared error.
print('Test set')
pred = linreg.predict(X_test)
print('Linear Regression mse: {}'.format(mean_squared_error(y_test, pred)))
print()
```

```python
from sklearn.metrics import r2_score

# Calculate R-squared for the training set
train_r2 = r2_score(y_train, linreg.predict(X_train))
print('R-squared for the Training set:', train_r2)

# Calculate R-squared for the test set
test_r2 = r2_score(y_test, linreg.predict(X_test))
print('R-squared for the Test set:', test_r2)
```
